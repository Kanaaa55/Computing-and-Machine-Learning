1. Aesthetics: Position, color, size, text; Variables: internet_users, gdp, four_regions, income.

```python
import pandas as pd
data3 = pd.read_csv('/Users/ruojiakuang/Desktop/dataset/q3data.csv')
data3 = data3.dropna(subset=['four_regions'])
print(data3)
```

```python
breaks_values = [200, 500, 1000, 2000, 5000, 10000, 20000, 50000, 100000]
labels_values = ['200', '500', '1000', '2000', '5000', '10k', '20k', '50k', '100k']

(ggplot(data3,
aes(
  x = "internet_users",
  y = "gdp",
  fill = "four_regions",
  size = "income"
))
+ annotate("text", x = 50, y = 3000, label = "2001", size = 200, colour = "grey", alpha = 0.5)
+ annotate("text", x = 85, y = 200, label = "\% of population", size = 10, colour = "black")
+ annotate("text", x = 0, y = 40000, label = "US$, inflation-adjusted", size = 10, colour = "black", angle = 90)
+ geom_point(alpha = 0.6)
+ scale_y_log10(breaks=breaks_values, labels=labels_values, limits=[0, 100000])
+ scale_x_continuous(breaks=[10, 20, 30, 40, 50, 60, 70, 80,90], limits=[0,90])
+ scale_size_continuous(range = (1, 30))
+ scale_fill_manual(values=['Cyan', 'lightgreen', 'lightcoral', 'Gold'])
+ xlab('Individuals using the Internet')
+ ylab('GDP/capita')
+ theme_linedraw()
+ theme(figure_size=(12, 6),)
)
```

```python
from plotnine import facet_wrap
(ggplot(data3,
aes(
  x = "internet_users",
  y = "gdp"
))
+ geom_boxplot()
+ facet_wrap("four_regions")
)
```

4. The boxplot with facet wrap is overall good for visualizing these data. It shows the distribution of gdp and internet users by each region. But the scatter plot does a better job by showing all four variables in one plot.
